```python
from keras.datasets import mnist
(train_images, train_lable),(test_images, test_lable) = mnist.load_data()
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
```

```python
digit = train_images[100]
plt.imshow(digit, cmap = plt.cm.binary)
plt.show()
print("so: ", train_lable[100])
```

```python
for i in range(15):
  plt.subplot(3,5,i+1)
  plt.tight_layout()
  plt.imshow(train_images[i], cmap = 'gray', interpolation='none')
  plt.title("so: {}".format(train_lable[i]) )
plt.xticks([])
plt.yticks([])
```

```python
model = Sequential()
model.add(Dense(512, activation = 'relu', input_shape = (28*28,)))
model.add(Dense(10,activation = 'softmax'))

model.summary()

X = test_images
Y = test_lable
```

```python
train_images=train_images.reshape(60000,28*28)
train_images = train_images.astype('float32')/255
test_images=test_images.reshape(10000,28*28)
test_images = test_images.astype('float32')
```

```python
from keras.utils import to_categorical

train_lable = to_categorical(train_lable, 10)
test_lable = to_categorical(test_lable, 10)

train_lable.shape
test_lable.shape
```

```python
# data
```

```python
model.compile(optimizer='rmsprop', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model.fit(train_images, train_lable, epochs=10, batch_size=128)
test_loss, test_acc = model.evaluate(test_images, test_lable)
print("loss:", test_loss)
print("acc:", test_acc)

```
