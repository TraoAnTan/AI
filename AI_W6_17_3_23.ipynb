```python
import pandas as pd
url = 'mtcars.csv'
data = pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
# moi tuong quan giua cac bien
correlation = data.corr()
print(correlation)

import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X= pd.DataFrame(data["wt"]), y = data["mpg"])
print("he so intercegt la", model.intercept_)
print("he so coffent la:", model.coef_)
model.score(X= pd.DataFrame(data["wt"]), y = data["mpg"])

y_prec = model.predict(X= pd.DataFrame(data["wt"]))
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg',figsize = (9,9),
                 color = 'black', xlim = (0,7))
plt.plot(data['wt'], y_prec, color = 'blue')
```

```python
# bac 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2,data['wt']**3 ]).T
model.fit(X, y =data['mpg'])
print('Intercept cua mo hinh:', model.intercept_)
print("he so coffent la:", model.coef_)
print("do chinh xac cua mo hinh:", model.score(X,y =data['mpg']))


```

```python
import numpy as np
range = np.arange(1.5,5.5,0.1)
X_range = pd.DataFrame([range, range**2, range**3]).T
y_prec = model.predict(X=X_range)
data.plot(kind = 'scatter', x= 'wt', y = 'mpg', 
               figsize = (9,9), color = 'black',xlim = (0,7))
plt.plot(range, y_prec, color = 'blue')
```

```python
# 50_Startups
import pandas as pd
url = '50_Startups.csv'
data = pd.read_csv(url)
data.head()
import matplotlib.pyplot as plt
data.hist()
plt.show()

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X= pd.DataFrame(data["Profit"]), y = data["Administration"])
print("he so intercegt la", model.intercept_)
print("he so coffent la:", model.coef_)
model.score(X= pd.DataFrame(data["Profit"]), y = data["Administration"])

correlation = data.corr()
print(correlation)

import seaborn as sns
sns.heatmap(correlation, annot = True)

y_prec = model.predict(X= pd.DataFrame(data["Profit"]))



data.plot(kind = 'scatter', x = 'R&D Spend', y = 'Profit',figsize = (9,9),
                 color = 'black', xlim = (0,7))
plt.plot(data['R&D Spend'], y_prec, color = 'blue')
```
